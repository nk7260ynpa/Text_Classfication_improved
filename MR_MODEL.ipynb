```python
import pandas as pd
import numpy as np
import io
#io 解決語法問題
```

```python
#讀進 pos資料
raw_data_pos = []
with io.open('data/MR/rt-polaritydata/rt-polarity.pos', encoding="latin-1") as fp:
    raw_data_pos = fp.readlines()
```

```python
#讀進 neg資料
raw_data_neg = []
with io.open('data/MR/rt-polaritydata/rt-polarity.neg', encoding="latin-1") as fp:
    raw_data_neg = fp.readlines()
```

```python
#去除空白
for i in range(len(raw_data_pos)):
    raw_data_pos[i] = raw_data_pos[i][0:-3]
for i in range(len(raw_data_neg)):
    raw_data_neg[i] = raw_data_neg[i][0:-3]
```

```python
#消除其他的符號
for i in range(len(raw_data_pos)):
    new = ''
    for s in raw_data_pos[i]:
        if s.isalpha():
            new += s
        if s == ' ':
            new += s
        raw_data_pos[i] = new
for i in range(len(raw_data_neg)):
    new = ''
    for s in raw_data_pos[i]:
        if s.isalpha():
            new += s
        if s == ' ':
            new += s
        raw_data_neg[i] = new
```

```python
df = pd.DataFrame({'text': raw_data_pos + raw_data_neg})
```

```python
df['target'] = 1
df['target'][5331:] = 0
```
